Q1 Find sum of first 10 natural numbers.

```python
sum=0
for i in range (1,11):
   sum=sum+i
print ("sum is:", sum)
```

Q2 Find factorial of a number.

```python
num=int(input("enter a number: "))
fact=1
for i in range(1,num+1):
   fact=fact*i
print("Factorial of given number is:", fact)
```

Q3 Print fibonacci series.

Q4 Find largest among 3 numbers.

```python
a=int(input("enter a number a:" ))
b=int(input("enter a number b:" ))
c=int(input("enter a number c:" ))
if (a>b & a>c):
   print ("a is the greatest:", a)
elif (b>a & b>c):
   print ("b is the greatest:", b)
else:
   print("c is the greatest:", c)
```

Q5 Create student result system.
input student details,input marks,calculate percentage,display grade,use if-elif,loops.

```python
#Taking student details.
name=input("enter student's name: ")
roll_number=input("enter student's roll number: ")

#Marks in various subjects.
marks_in_subject1=float(input("enter marks of subject 1: "))
marks_in_subject2=float(input("enter marks of subject 2: "))
marks_in_subject3=float(input("enter marks of subject 3: "))

#Total marks scored
total_marks=(marks_in_subject1+marks_in_subject2+marks_in_subject3)

#Percentage scored
percentage=(total_marks/300)*100

#Grading system
if percentage >= 93:
      grade="A+"
elif percentage >= 85:
       grade="A"
elif percentage >= 77:
       grade="B"
else:
       grade="C"

#Showcasing  the report.
print("\n    STUDENT REPORT   ")
print("Name:", name)
print("Roll number:",roll_number )
print("Total marks obtained:", total_marks)
print("Percentage:", percentage,"%")
print("Grade:", grade)
```
