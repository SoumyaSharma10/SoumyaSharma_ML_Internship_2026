Q1 Find area of rectangle.

Q2 Find simple interest.

```python
principal=float(input("enter the principal amount:" ))
rate=float(input("enter rate of interest:" ))
time=float(input("enter time period in years:" ))
Simple_Interest=(principal*rate*time)/100
print("Simple Interest is:", Simple_Interest)
```

Q3 Convert temperature from celsius to fahrenheit.

```python
celsius=float(input("enter temp. in celsius:" ))
fahrenheit=(celsius*9/5)+32
print("Temp. in fahrenheit:", fahrenheit)
```

Q5 Find square and cube of a number.

```python
number=int(input("enter a number:" ))
square=(number*number)
cube=(number*number*number)
print("square of given number is:", square)
print("cube of given number is:", cube)
```

Q7 Create a student report program that takes student details using input(),stores marks in variables. calculates total percentage.Use proper comments and proper indentation.

```python
#Taking student details.
name=input("enter student's name: ")
roll_number=input("enter student's roll number: ")

#Marks in various subjects.
marks_in_subject1=float(input("enter marks of subject 1: "))
marks_in_subject2=float(input("enter marks of subject 2: "))
marks_in_subject3=float(input("enter marks of subject 3: "))

#Total marks scored
total_marks=(marks_in_subject1+marks_in_subject2+marks_in_subject3)

#Percentage scored
percentage=(total_marks/300)*100

#Grading system
if percentage >= 93:
      print(grade="A+")
elif percentage >= 85:
       grade="A"
elif percentage >= 77:
       grade="B"
else:
       grade="C"

#Showcasing  the report.
print("STUDENT REPORT")
print("Name:", name)
print("Roll number:",roll_number )
print("Total marks obtained:", total_marks)
print("Percentage:", percentage)
print("Grade:", grade)

```
